```python
#Create three methods in each class, 2 methods are specific to each class and third 
#method (override method) should be in all three Classes A, B ,c
class A:
    def methodA1(self):
        print("Method A1 in class A")

    def methodA2(self):
        print("Method A2 in class A")

    def overriddenMethod(self):
        print("Overridden method in class A")
class B(A):
    def methodB1(self):
        print("Method B1 in class B")

    def methodB2(self):
        print("Method B2 in class B")

    def overriddenMethod(self):
        print("Overridden method in class B")
class C(B):
    def methodC1(self):
        print("Method C1 in class C")

    def methodC2(self):
        print("Method C2 in class C")

    def overriddenMethod(self):
        print("Overridden method in class C")
def main():
    objA = A()
    objB = B()
    objC = C()
    
    objA.methodA1()
    objA.methodA2()
    objA.overriddenMethod()

    objB.methodB1()
    objB.methodB2()
    objB.overriddenMethod()  

    objC.methodC1()
    objC.methodC2()
    objC.overriddenMethod()  
    refB = B()
    refC = C()

    refB.overriddenMethod()  
    refC.overriddenMethod() 
if __name__ == "__main__":
    main()
```

```python
#Create a class with main method. Create an object for each class A, B and C in main 
#method and call every method of each class using its own object/instance.
class A:
    def __init__(self):
        self.instanceVarA = "Instance variable in A"

    def methodA1(self):
        print("Method A1 in class A")

    def methodA2(self):
        print("Method A2 in class A")

    def overriddenMethod(self):
        print("Overridden method in class A")
class B(A):
    def __init__(self):
        super().__init__()
        self.instanceVarB = "Instance variable in B"

    def methodB1(self):
        print("Method B1 in class B")

    def methodB2(self):
        print("Method B2 in class B")

    def overriddenMethod(self):
        print("Overridden method in class B")
class C(B):
    def __init__(self):
        super().__init__()
        self.instanceVarC = "Instance variable in C"

    def methodC1(self):
        print("Method C1 in class C")

    def methodC2(self):
        print("Method C2 in class C")

    def overriddenMethod(self):
        print("Overridden method in class C")
def main():
    objA = A()
    objB = B()
    objC = C()

    print("objA instance variable:", objA.instanceVarA)
    print("objB instance variable:", objB.instanceVarB)
    print("objC instance variable:", objC.instanceVarC)
    refB = B()
    refC = C()
    print("refB instance variable:", refB.instanceVarB)
    print("refC instance variable:", refC.instanceVarC)

if __name__ == "__main__":
    main()
```

```python
#Runtime Polymorphism with Data Members/Instance variables, Repeat the above 
#process only for data members
class A:
    def __init__(self):
        self.instanceVarA = "Instance variable in A"  

    def methodA1(self):
        print("Method A1 in class A")

    def methodA2(self):
        print("Method A2 in class A")

    def overriddenMethod(self):
        print("Overridden method in class A")


class B(A):
    def __init__(self):
        super().__init__()
        self.instanceVarB = "Instance variable in B"  

    def methodB1(self):
        print("Method B1 in class B")

    def methodB2(self):
        print("Method B2 in class B")

    def overriddenMethod(self):
        print("Overridden method in class B")


class C(B):
    def __init__(self):
        super().__init__()
        self.instanceVarC = "Instance variable in C" 

    def methodC1(self):
        print("Method C1 in class C")

    def methodC2(self):
        print("Method C2 in class C")

    def overriddenMethod(self):
        print("Overridden method in class C")


def main():
    refB = B()  
    refC = C()  
    print("refB instanceVarA:", refB.instanceVarA)  
    print("refB instanceVarB:", refB.instanceVarB)  

    print("refC instanceVarA:", refC.instanceVarA)  
    print("refC instanceVarB:", refC.instanceVarB)  
    print("refC instanceVarC:", refC.instanceVarC)  

    print("\nAccessing instance variables via superclass references:")
    print("refB instanceVarA:", refB.instanceVarA)  
    print("refC instanceVarA:", refC.instanceVarA)  
    print("refC instanceVarB:", refC.instanceVarB)  
    print("refC instanceVarC:", refC.instanceVarC) 


if __name__ == "__main__":
    main()
```
